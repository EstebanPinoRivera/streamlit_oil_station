```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import json
import time
import random
```

```python
API_KEY = '' #API HERE Platform
```

```python
#'https://preciobencina.cl/bencineras-en-comuna-de-losangeles.php'
#'https://preciobencina.cl/bencineras-en-region-del-biobio.php'
url = requests.get('https://preciobencina.cl/bencineras-en-comuna-de-losangeles.php')
soup = BeautifulSoup(url.content, 'html.parser')
```

<h3>Get station and address<h3>

```python
station_name = []
elements_a = soup.find_all('a', class_="job-listings")
for elements_h3 in elements_a:
    elements_a = elements_h3.find('h3')
    if elements_a:
        station_name.append(elements_a.text.strip())
station_name
```

<h3>Get address name<h3>

```python
data = {'Station_name': [], 'Station_address': []}

elements_a = soup.find_all('a', class_="job-listings")

for element_a in elements_a:
    name_element = element_a.find('h3')
    address_element = element_a.find('span', class_='company-neme')

    if name_element and address_element:
        data['Station_name'].append(name_element.text.strip())
        data['Station_address'].append(address_element.text.strip())

df = pd.DataFrame(data)
df['City'] = 'Los Ángeles'
df['Full_address'] = df['Station_address'] + ', '+ df['City'] + ', ' + 'Chile'
df
```

```python
df = pd.read_excel('stations.xlsx')
df.head(5)
```

```python
df2 = df[['Station_name', 'Station_address', 'Full_address']].drop_duplicates(subset= 'Station_name')
df2.reset_index(inplace= True)
df2.drop(columns= 'index', inplace= True)
df2.head(5)
```

```python
address = df2['Full_address'][2]
url = f'https://geocode.search.hereapi.com/v1/geocode?q={address}&apiKey={API_KEY}'
response = requests.get(url).json()
```

```python
def Get_Coords(address, API_KEY):
    url = f'https://geocode.search.hereapi.com/v1/geocode?q={address}&apiKey={API_KEY}'
    
    try:
        response = requests.get(url).json()
        CleanAddress = response['items'][0]['title'].upper()
        LAT = response['items'][0]['position']['lat']
        LNG = response['items'][0]['position']['lng']
        
        results = [CleanAddress, LAT, LNG]
        
    except:
        
        results = ['Not found', 'NA', 'NA']
        
    return results
```

```python
Get_Coords(address, API_KEY)
```

<h3>Get coords<h3>

```python
coords = []
times = [1, 2, 2,5]

for i in range(len(df2['Full_address'])):
    
    try:
        R = Get_Coords(df2['Full_address'][i], API_KEY)
        geo_source = R[1], R[2]
        coords.append(geo_source)
    except:
        coords.append('Error')
    
    time.sleep(random.choices(times)[0])
```

```python
df2['Coords'] = coords
df2.head(5)
```

```python
df_final = pd.merge(df, df2 [['Full_address', 'Coords']], how= 'inner', on= 'Full_address')
df_final.head(5)
```
